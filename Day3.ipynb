DIFFERENT TYPES OF DATA

1. Categorical data

* Categorical data is the statistical data comprising categorical variables of the data that are converted into categories.

* Categorical variables represents types of data which may be divided into groups .examples of categorical data are race,age,sex etc.

* Analysis of categorical data generally involves the use of data tables.A two way table presents categorical data by counting the number of observations that fall into each group for two variables ,one divided into rows and the other divided into columns.

* Bar charts and pie charts are great tools for comparing two or more categorical values against each other .They just represents the number of things in the category.

2.Continous data

* Continous data refers to variables that can take on any values within specific range.example :-age,weight,height,length etc.

* The data is continous if its value is being applied to somtheing that can be counted ,not necessarily by whole numbers ,but by fractions of itself.

* These variables are typically measured using numerical values ,and they can be further subdivided into interval and ratio data.

3. Discrete data



* Discrete data is a type of quantitive data that includes nondivisible figures and statistics you can count.

* Typically the discrete datas are written as numbers that represents the exact values.

* Discrete data is distinct,individual,unique numbers like counting.The attendance at a soccer game is an example for discrete data.

* Discrete data can be summarized with the frequency table that shows the count of each category in the data sample.

4.Nominal data

* Nominal data is qualitative data used to name or label variables without providing numeric values

* It is the most straightforward types of measurement scale.Nominal variables are labeled into categories that do not overlap.

* Data is not measured or evalvated in this case;it is just assigned to multiple groups .these groups are unique and have no common element.
